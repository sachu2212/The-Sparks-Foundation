Rename Important Columns values

```python
data.rename(columns={'iyear':'Year','imonth':'Month','iday':"Day",'gname':'Group','country_txt':'Country',
                     'region_txt':'Region','provstate':'Province','city':'City','latitude':'Latitude',
                     'longitude':'Longitude','summary':'Summary','attacktype1_txt':'Attacktype',
                     'targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'Killed',
                     'nwound':'Wounded'},inplace = True)
```

```python
data = data[['Year','Month','Day','Country','Province','Region','City','Latitude','Longitude',"Attacktype",
             'Killed','Wounded','target1','Summary','Group','Targettype','Weapon','motive']]
data.head()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(data.corr(), annot= True)
plt.show()
```

```python
print('Country With Highest Terror Attacks: ', data["Country"].value_counts().idxmax())
print('City With Highest Terror Attacks: ',data["City"].value_counts().index[1])
print('Region With Highest Terror Attacks: ',data["Region"].value_counts().idxmax())
print('Year With Highest Terror Attacks: ',data["Year"].value_counts().idxmax())
print('Group with highest terror attacks: ',data["Group"].value_counts().index[1])
print('Type Of Attack With Highest Terror Attacks: ',data["Attacktype"].value_counts().idxmax())
```

```python
#Bar Plot
sns.set_style("darkgrid")
year = data['Year'].unique()
years_count = data['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (15,10))
sns.barplot(x = year, y = years_count, palette = 'hsv')
plt.xticks(rotation = 90)
plt.xlabel('Attacking Year', fontsize=25)
plt.ylabel('Number of Attacks Each Year', fontsize=25)
plt.title('Attacks In Years', fontsize=25)
plt.show()
```

```python
data['Country'].value_counts()[:10]
```

```python
#Bar Plot
plt.subplots(figsize=(20,10))
country = data['Country'].unique()
countries_count = data['Country'].value_counts(dropna = False).sort_index()
sns.barplot(x=country, y=countries_count)
sns.barplot(data['Country'].value_counts()[:10].index,data['Country'].value_counts()[:10].values,palette='Dark2')
plt.title("Top 10 Countries With Highest Terror Attacks", fontsize=25)
plt.xlabel('Name Of Countries', fontsize=25)
plt.ylabel('Counts', fontsize=25)
plt.xticks(rotation=90)
plt.show()
```

```python
#Bar Plot
plt.subplots(figsize=(20,10))
sns.barplot(data['City'].value_counts()[:10].index,data['City'].value_counts()[:10].values,palette= 'nipy_spectral')
plt.title("Visualising Top 10 Cities With Highest Terror Attacks", fontsize=25)
plt.xlabel('Cities', fontsize=25)
plt.ylabel('Count', fontsize=25)
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.subplots(figsize=(12,7))
sns.kdeplot(data['Year'], hue=data['Region'])
plt.title('Terrorist Activities By Region In Each Year', fontsize=25)
plt.xlabel('Year', fontsize=25)
plt.ylabel('Frequency Of Attacks', fontsize=25)
plt.xticks(rotation = 90)
plt.show()
```

```python
#Bar Plot 
plt.subplots(figsize=(12,7))
sns.barplot(data['Province'].value_counts()[:10].index, data['Province'].value_counts()[:10].values, palette='colorblind')
plt.title("Visualising Top 10 Province With Highest Terror Attacks", fontsize=25)
plt.xlabel('Province', fontsize=25)
plt.ylabel('Number Of Attacks', fontsize=25)
plt.xticks(rotation = 90)
plt.show()
```

```python
df=data[['Attacktype','Killed']]
df=df.groupby(['Attacktype','Killed'],axis=0).sum().sort_values('Killed',ascending=False).drop('Unknown').reset_index().head(10)
df
```

```python
#Bar Plot
plt.subplots(figsize=(12,7))
data.groupby(['Attacktype'])['Killed'].sum().sort_values(ascending = False).head(20).plot(kind = 'bar', colormap = 'summer')
plt.title("Number Of People Killed In Each Attack Type", fontsize=25)
plt.xlabel('Attack Type', fontsize=25)
plt.ylabel('People killed', fontsize=25)
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.subplots(figsize=(12,7))
data.groupby(['Year'])['Killed'].sum().sort_values(ascending = False).head(20).plot(kind = 'bar', colormap = 'Dark2')
plt.title("Visualising Total Deaths In Year", fontsize=25)
plt.xlabel('Attack Year', fontsize=25)
plt.ylabel('Number Of Deaths', fontsize=25)
plt.xticks(rotation = 90)
plt.show()
```

```python
df=data[['Region','Group','Killed']]
df=df.groupby(['Group','Region'],axis=0).sum().sort_values('Killed',ascending=False).drop('Unknown').reset_index().head(10)
df
```

```python
count_terror = data['Region'].value_counts()[:10].to_frame()
count_terror.columns=['Wounded']
count_kill=data.groupby ('Region')['Killed'].sum().to_frame()
count_terror.merge(count_kill,left_index = True,right_index =True,how='left').plot.bar(width=0.9)
fig=plt.gcf()
plt.title("Regions Having Wounded And Killed People ", fontsize=20)
plt.xlabel('Region', fontsize=20)
plt.ylabel('Number of Wounded And Killed People', fontsize=20)
fig.set_size_inches(12,7)
plt.show()
```

```python
df=data[['Country','Group','Killed']]
df=df.groupby(['Group','Country'],axis=0).sum().sort_values('Killed',ascending=False).drop('Unknown').reset_index().head(10)
df
```

```python
count_terror = data['Country'].value_counts()[:10].to_frame()
count_terror.columns=['Wounded']
count_kill=data.groupby ('Country')['Killed'].sum().to_frame()
count_terror.merge(count_kill,left_index = True,right_index =True,how='left').plot.bar(width=0.9)
fig=plt.gcf()
plt.title("Countries Having Wounded And Killed People ", fontsize=20)
plt.xlabel('Country', fontsize=20)
plt.ylabel('Number of Wounded And Killed People', fontsize=20)
fig.set_size_inches(12,7)
plt.show()
```

```python
df=data[['City','Group','Killed']]
df=df.groupby(['Group','City'],axis=0).sum().sort_values('Killed',ascending=False).drop('Unknown').reset_index().head(10)
df
```

```python
count_terror = data['City'].value_counts()[:10].to_frame()
count_terror.columns=['Wounded']
count_kill=data.groupby ('City')['Killed'].sum().to_frame()
count_terror.merge(count_kill,left_index = True,right_index =True,how='left').plot.bar(width=0.9)
fig=plt.gcf()
plt.title("Cities Having Wounded And Killed People ", fontsize=20)
plt.xlabel('City', fontsize=20)
plt.ylabel('Number of Wounded And Killed People', fontsize=20)
fig.set_size_inches(12,7)
plt.show()
```

```python
plt.figure(figsize = (12,7))
sns.barplot(x = data['Group'].value_counts()[1:11].values, y = data['Group'].value_counts()[1:11].index,palette='magma')
plt.title('Top 10 Terrorist Organisation with Highest Terror Attacks', fontsize=25)
plt.ylabel('Terror Group Name', fontsize=25)
plt.xlabel('Number of Attacks', fontsize=25)
plt.xticks(rotation= 90)
plt.show()
```
